**Part 2. Convert the general format and data structure of the data source and take user input to choose file format**
*   CSV to JSON to SQL
*   JSON to CSV to SQL

```python
#2a: csv to json to sql

import pandas as pd
import sqlite3
dfa = pd.read_csv('worldwide_crop_consumption.csv')

# user input to choose file format (csv/json/sql)
answer= input()
if answer == 'csv':
    dfa.to_csv('worldwide_crop_consumption.csv')
    display(dfa)
elif answer == 'json':
    dfa.to_json('worldwide_crop_consumption.json')
    display(dfa)
elif answer == 'sql':
    dfa.to_sql('worldwide_crop_consumption.sql', con=sqlite3.connect('worldwide_crop_consumption.sql'))
    print(dfa)
else:
    print('Invalid input')
```

```python
#2b: json to csv to sql

import pandas as pd
import sqlite3
# Add 'lines=True' to handle JSON with multiple objects
dfb = pd.read_json('label_num_to_disease_map.json', lines=True)

# user input to choose file format (csv/json/sql)
answer= input()
if answer == 'csv':
    dfb.to_csv('label_num_to_disease_map.json', index=False) # Add index=False to avoid writing index
    display(dfb)
elif answer == 'json':
    dfb.to_json('label_num_to_disease_map.json', orient='records', lines=True) # Use orient='records' and lines=True for consistent output
    display(dfb)
elif answer == 'sql':
    dfb.to_sql('label_num_to_disease_map.json', con=sqlite3.connect('label_num_to_disease_map.sql'))
else:
    print('Invalid input')
```

```python
#3b: make changes to json , add column

def modify_columns(dfb):

    #5: brief summary of the data file - rows and columns
    print(f"Original Columns: {dfb.columns.tolist()}")
    print(f"Pre-modification: {dfb.shape[0]} rows, {dfb.shape[1]} columns")

    try:
        dfb['5'] = 'Bacterial Leaf Blight of Rice (BLB)'
        print("Added '5' column with default value 'Bacterial Leaf Blight of Rice (BLB)'.")
    except Exception as e:
        print(f"Error: {e}")

    #6: brief summary of the data post-processing - rows and columns
    print(f"Modified Columns: {list(dfb.columns)}")
    print(f"Post-modification: {dfb.shape[0]} rows, {dfb.shape[1]} columns\n")
    return dfb

dfb = modify_columns(dfb)
dfb.head()
```
